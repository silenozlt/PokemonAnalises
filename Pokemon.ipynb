# Pré-Processamento

```python
pokemon_df.drop(["japanese_name","pokedex_number","base_egg_steps","classfication","percentage_male"], axis=1, inplace=True)
```

```python
pokemon_df.rename(str.title, axis="columns", inplace=True)

types = [t for t in list(pokemon_df["Type1"].unique())]

pokemon_df["Type1"].replace(type, [t.title() for t in types], inplace=True)
pokemon_df["Type2"].replace(type, [t.title() for t in types], inplace=True)
```

```python
pokemon_df["Type"] = pokemon_df.apply(
    lambda x: x["Type1"] if pd.isnull(x["Type2"]) else f'{x["Type1"]}_{x["Type2"]}', axis=1)
```

```python
 pokemon_df["Abilities"]  = pokemon_df.apply(lambda x: literal_eval(x["Abilities"]), axis=1)
```

```python
pokemon_df.head(10)
```

```python
pokemon_df["N_abilities"] = pokemon_df.apply(lambda x: len("Abilities"), axis=1)
```

```python
pokemon_df["IMC"] = pokemon_df.apply(lambda x: x["Weight_Kg"]/(x["Height_M"]**2), axis=1)
```

## Visualização de Dados

Quantos novos pokemons existem por geração ?

```python
generations = [f'Geração {g}' for g in pokemon_df.Generation.unique()]

ax = sns.countplot(y=pokemon_df.Generation, orient="v", color="blue")

ax.set_xlabel("")
ax.set_xticklabels([])

ax.set_ylabel("Geração")
ax.set_yticklabels(generations, size=30)

ax.set_title("Novos Pokemons por geração", fontdict={"color": "black"}, loc="center")

for index, value in enumerate(pokemon_df.Generation.value_counts().sort_index()):
    plt.annotate(f'{value}', xy=(value + 1, index), color="red")
```

## Quais são os tipos primarios mais existentes ?


```python
type1_value_count = pokemon_df["Type1"].value_counts()
```

```python
type1_value_count
```

```python
palette = ['#6991F0', '#A8AA79', '#7AC852', '#A7B822', '#F85887', '#EF812E', '#B99F38', '#F6D030','#A0429F',
           '#BCA23B','#6D5947','#C12F27','#70589A','#6B3EE3','#B6B8D0','#9AD7D9','#FF65D5', '#A991F0']

ax = squarify.plot(type1_value_count,
                  label=type1_value_count.index,
                  color=palette,
                  pad=0.1,
                  text_kwargs={"fontsize":30, "color":"black"})

ax.set_title("Quantidade de tipos primarios de Pokemons",fontdict={"color":"black"}, loc="center")
plt.axis("off");
```

## Quais os tipos secundarios mais existentes ?

```python
pokemon_df["Type2"].isnull().sum()
```

```python
type2_value_counts = pokemon_df["Type2"].value_counts()
types_df = pd.concat([type1_value_count, type2_value_counts], axis = 1)
types_df.sort_values(by="Type2", ascending=False, inplace=True)
```

```python
ax = types_df.plot.barh(stacked=True, color=["green","DeepSkyBlue"], width=0.8)

ax.invert_yaxis()

ax.get_legend().remove()

ax.annotate("Tipo 1", xy=(0,0), xytext=(0,-1), color="green", size=16)
ax.annotate("Tipo 2", xy=(7.5,0), xytext=(7.5,-1), color="DeepSkyBlue", size=16)

ax.set_xlabel(xlabel="Quantidade")

ax.set_yticklabels(ax.get_yticklabels(), size= 14)

ax.set_title("Pokemon por tipos primarios e secundarios", fontdict={"color":"black"}, loc="center")

sns.despine(top=True, right=True, left=True, bottom=True)

for index, row in enumerate(types_df.iterrows()):
  plt.annotate(f'{row[1]["Type1"]}',xy=(1, index), color="black")
  plt.annotate(f'{row[1]["Type2"]}',xy=(row[1]["Type1"] + 2, index), color="black")
```

Quais são as combinações de tipo mais comuns



```python
top10 = pokemon_df[~pokemon_df["Type2"].isnull()]["Type"].value_counts()[:10]
```

```python
top10
```

```python
ax = sns.barplot(y=top10.index, x=top10.values, orient="h", color="DarkViolet")

ax.set_xlabel("")
ax.set_xticklabels([])

ax.set_yticklabels(ax.get_yticklabels(), size=14)

ax.set_title("Frequencia de combinaçõies de tipos de pokemons", fontdict={"color": "black"}, loc="center")

for index, value in enumerate(top10):
    plt.annotate(f'{value}', xy=(value + 0.1, index), color="black")
```
